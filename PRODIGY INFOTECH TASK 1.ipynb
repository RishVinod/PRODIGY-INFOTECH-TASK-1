```python
plt.figure(figsize=(16, 10)) 
plt.bar(df.State, df.Population, color='grey')


plt.title('Population of States')
plt.xlabel('States')
plt.ylabel('Population')

plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(16, 10)) 
plt.bar(df.State, df.Male, color='grey')


plt.title('Statewise Male Population')
plt.xlabel('State')
plt.ylabel('Male')

plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```

```python
plt.figure(figsize=(16, 10)) 
plt.bar(df.State, df.Female, color='grey')


plt.title('Statewise Female Population')
plt.xlabel('State')
plt.ylabel('Female')

plt.xticks(rotation=90)
plt.tight_layout()

plt.show()
```
